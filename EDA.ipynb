```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
data= load_breast_cancer()
df=pd.DataFrame(data.data,columns=data.feature_names)
df['target']=data.target
df.head()
```

```python
sns.histplot(data=df, x='mean radius', hue='target', kde=True)
plt.show()
```

The distribution shows that values up to around 12 are predominantly associated with benign tumors (class 1), while higher values are increasingly dominated by malignant tumors (class 0). This indicates a clear shift in class distribution as the feature value increases. Although some overlap exists, the feature demonstrates good separation and strong predictive capability for classification.
